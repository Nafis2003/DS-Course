<h1 style='font-weight:600;text-align:center;'>Data Visualization With Seaborn</h1>

<p>Lung cancer is one of the leading causes of cancer-related deaths worldwide, and its prevalence is influenced by multiple factors, including smoking habits, air pollution, healthcare infrastructure, and economic conditions. This dataset provides a comprehensive analysis of lung cancer risk and trends across 25 countries, helping researchers and analysts identify patterns, correlations, and potential causes behind varying lung cancer rates. By examining factors such as smoking prevalence, mortality rates, and GDP per capita, this dataset can be used for trend forecasting, policy recommendations, and healthcare research.
<br>
Dataset link : <a>https://www.kaggle.com/datasets/ankushpanday1/lung-cancer-risk-and-trends-across-25-countries</a>
</p>


```python
#importing libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#DataFrame
df=pd.read_csv('lung_cancer_prediction_dataset.csv',index_col='ID')
df.drop_duplicates()
df.head(5)
```

```python
#Number of rows and columns
r,c=df.shape
print(f'The dataset has {r} rows and {c} columns')
```

<h3 style='font-weight:500'>Columns and Descriptions</h3>
<ul>
  <li><b>ID:</b> Unique identifier for each record.</li>
  <li><b>Country:</b> Name of the country.</li>
  <li><b>Population_Size:</b> Total population of the country.</li>
  <li><b>Age:</b> Age of the individual or group.</li>
  <li><b>Gender:</b> Gender of the individual </li>
  <li><b>Smoker:</b> Indicates whether the individual is a smoker.</li>
  <li><b>Years_of_Smoking:</b> Total years the individual has smoked.</li>
  <li><b>Cigarettes_per_Day:</b> Average number of cigarettes smoked daily.</li>
  <li><b>Passive_Smoker:</b> Indicates exposure to second-hand smoke.</li>
  <li><b>Family_History:</b> Presence of lung cancer in family history.</li>
  <li><b>Lung_Cancer_Diagnosis:</b> Diagnosis status of lung cancer.</li>
  <li><b>Cancer_Stage:</b> Stage of lung cancer.</li>
  <li><b>Survival_Years:</b> Number of years the patient has survived after diagnosis.</li>
  <li><b>Adenocarcinoma_Type:</b> Type of lung cancer subtype.</li>
  <li><b>Air_Pollution_Exposure:</b> Exposure level to air pollution.</li>
  <li><b>Occupational_Exposure:</b> Workplace exposure to carcinogens.</li>
  <li><b>Indoor_Pollution:</b> Level of pollution inside home.</li>
  <li><b>Healthcare_Access:</b> Access to quality healthcare.</li>
  <li><b>Early_Detection:</b> Whether the cancer was detected early.</li>
  <li><b>Treatment_Type:</b> Type of treatment received.</li>
  <li><b>Developed_or_Developing:</b> Classification of the country.</li>
  <li><b>Annual_Lung_Cancer_Deaths:</b> Number of annual deaths due to lung cancer.</li>
  <li><b>Lung_Cancer_Prevalence_Rate:</b> Percentage of population affected by lung cancer.</li>
  <li><b>Mortality_Rate:</b> Death rate due to lung cancer in the population.</li>
</ul>



<h3 style='font-weight:500'>Plots</h3>

```python
#Histogram
sns.set_theme(style='white')
plt.figure(figsize=(16,9))
sns.histplot(x='Age', data=df)
plt.title('Age Distribution')
plt.show()
```

<h4>Insight</h4>
<p>The age distribution reveals a relatively even spread across most age groups, indicating diverse representation in the dataset. However, distinct spikes at ages like 20, 60 suggest possible age rounding during data entry. This could slightly distort the actual age-based patterns. Overall, the dataset provides a broad demographic overview for analysis, but preprocessing may be needed to address the spikes.</p>

```python
#Pairplot
plt.figure(dpi=300)
ndf=df.select_dtypes(include=[np.number])
sns.pairplot(ndf, height=5,diag_kind='kde')
plt.suptitle('PairPlot',y=1.01)
plt.show()
```

<h4>Insights</h4>
<p>The pairplot provides a comparative view of numerical variables, highlighting their pairwise relationships. From the plot, some variables show strong linear relationships, such as between smoking-related variables and lung cancer prevalence. However, there are clusters and outliers in some combinations, indicating variations across different subgroups or countries.</p>

```python
#Heatmap
plt.figure(figsize=(16,9))
ndf=df.select_dtypes(include=[np.number])
sns.heatmap(ndf.corr(),annot=True)
plt.title('Heatmap')
plt.show()
```

<h4>Insights</h4>
<p>The heatmap of correlations among numeric variables shows that smoking-related factors like Years_of_Smoking and Cigarettes_per_Day are strongly correlated with lung cancer prevalence. Interestingly, Air_Pollution_Exposure also has a moderate correlation with mortality rates, suggesting environmental factors play a role. Weak or negligible correlations are seen with variables like Population_Size and Gender, which may have minimal direct influence.</p>

```python
#JointPlot
plt.figure(figsize=(16,9))
sns.jointplot(x='Age',y='Cigarettes_per_Day',data=df[df.Smoker=='Yes'],fill=True,kind='kde')
plt.show()
```

<h4>Insights</h4>
<p>The jointplot demonstrates a clear trend where older individuals tend to smoke more cigarettes daily compared to younger smokers. A density cluster around middle-aged individuals indicates a high concentration of smokers in that age group, potentially reflecting cultural or generational smoking habits.</p>

```python
#Barplot
plt.figure(figsize=(16,9))
sns.barplot(x='Country',y='Age',hue='Gender',data=df,estimator=np.mean,errorbar=None)
plt.xticks(rotation=90)
plt.show()
```

<h4>Insights</h4>
<p>The barplot highlights the average age distribution across countries, segregated by gender. In most countries, males appear to have a slightly higher average age compared to females. However, variations across countries suggest differences in population demographics or sampling methods.</p>

```python
#Violin Plot
plt.figure(figsize=(16,9))
sns.violinplot(data=df, x="Developed_or_Developing",y="Lung_Cancer_Prevalence_Rate")
plt.show()
```

<h4>Insights</h4>
<p>The violin plot reveals that lung cancer prevalence rates are generally higher in developed countries compared to developing ones. However, the distribution in developing countries shows greater variability, suggesting that specific factors within those nations might amplify or mitigate risks.</p>
